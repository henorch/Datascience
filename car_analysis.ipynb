```python
import pandas as pd
import matplotlib as plt
from bokeh.plotting import figure,show
from bokeh.models import HoverTool
df = pd.read_csv("./car_price.csv")
df
```

```python
#checking out the datatype
main = df[["CarName", "enginetype", "price", "cylindernumber", "horsepower"]]
main
```

```python
mean_price = df["price"].mean()
mean_price
```

```python
main[main["cylindernumber"].isin(["four", "five"])]
```

```python
ab_av = main.loc[main["price"] > mean_price, ["CarName", "enginetype", "cylindernumber", "horsepower", "price"]]
ab_av
```

```python
ab_av["diff"] = ab_av["price"] - mean_price
ab_av
```

```python
ab_av.groupby("enginetype").mean()
```

```python
TOOLTIP = HoverTool(tooltips = [
    ("index", "$index"),
    ("(horsepower, price)", "(@horsepower, @price)"),
    ("car name", "@CarName")
])

p = figure(title="car analysis", x_axis_label = "price", y_axis_label="horsepower", width=700, height=700, tools=[TOOLTIP])

p.circle("price", "horsepower", size=5, source=main)
show(p)
```
